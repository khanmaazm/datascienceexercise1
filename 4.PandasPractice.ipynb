# Simple Joins and Concatenate and Merge

```python
df1 = pd.DataFrame({'numbers': [1, 2, 3,5], 'colors': ['red', 'white', 'blue','black'], 'colorsdesc': ['One of the Base Color','Most used color','Sky is blue','Opposite of White']})
df1
```

```python
df2 = pd.DataFrame({
                    'numbers': [1, 2, 3,4],
                    'colors': ['red', 'white', 'blue','green']
                    })
print(type(df2))
df2.head()
```

```python
#join is exactly like SQL joins 
#on which column to join is by default index-on-index
joindf = df1.join(df2, how = 'inner', lsuffix = 'adv')
joindf.head()
```

```python
#join is exactly like SQL joins
#on = Column or index in the caller to join on the index in others
joindf = df1.join(df2, on=['numbers'], how = 'inner', lsuffix = 'adv')
joindf.head()
```

```python
#Rowwise Contactenation
rowconcat = pd.concat([df1,df2],axis=0)
rowconcat.head()
```

```python
#Columnwise Contactenation.Similar to Join
pd.concat([df1,df2],axis=1)
```

```python
mergeinner = pd.merge(df1,df2, on='numbers', how='inner')
mergeinner
```

```python
mergeouter = pd.merge(df1,df2, on='numbers', how='outer')
mergeouter
```

```python
mergeleft = pd.merge(df1,df2, on='numbers', how='left')
mergeleft
```

```python
mergeright = pd.merge(df1,df2, on='numbers', how='right')
mergeright
```

```python
mergemultiplecols = pd.merge(df1,df2, on=['numbers','colors'], how='right')
mergemultiplecols
```
