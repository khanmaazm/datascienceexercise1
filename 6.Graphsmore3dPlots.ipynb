```python
import matplotlib.pyplot as plt
import numpy as np
```

# https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html

```python
x = np.random.normal(20,4,100)
n = np.random.normal(0, 250, 100)
xsq = x * x
z = 14 * xsq + 8 * x + 1300 + n
xdectree = np.random.normal(20,4,100)
ndectree = np.random.normal(0, 100000, 100)
ypoly2dectree = 67*xdectree*xdectree*xdectree - 0.25*xdectree*xdectree + 40*xdectree + 3790
zpoly2dectree = ypoly2dectree + ndectree
plt.figure(figsize=(5,5))
plt.scatter(xdectree, zpoly2dectree)
plt.figure(figsize=(5,5))
plt.scatter(x, z)
```

```python
ax = plt.axes(projection='3d')
#ax.plot3D(x, xsq, z)
ax.scatter3D(x, xsq, z);
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3d contour')
```

```python
ax = plt.axes(projection='3d')
#ax.plot3D(x, xsq, z)
#ax.scatter3D(xdectree, ypoly2dectree, zpoly2dectree);
ax.plot3D(xdectree, ypoly2dectree, zpoly2dectree);
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3d contour')
```

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
xr = np.ravel(r * np.sin(theta))
yr = np.ravel(r * np.cos(theta))
f = lambda x, y: np.sin(np.sqrt(x ** 2 + y ** 2))
zr = f(xr, yr)
ax = plt.axes(projection='3d')
ax.scatter(xr, yr, zr, c=zr, cmap='viridis', linewidth=0.5);    
```

```python
ax = plt.axes(projection='3d')
ax.plot_trisurf(xr, yr, zr, cmap='viridis', edgecolor='none');
```
